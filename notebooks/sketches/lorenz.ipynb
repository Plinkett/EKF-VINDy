```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Lorenz chaotic ODE

Original Lorenz system

$$
\begin{align}
\dot{z}_1&=\sigma(z_2-z_1),\\
\dot{z}_2&=z_1(\rho - z_3)-z_2,\\
\dot{z}_3&=z_1 z_2-\beta z_3.
\end{align}
$$

We define the ODE...

```python
params = {
    'sigma': 10,
    'rho': 50,
    'beta': 8/3
}

def lorenz(z, t, params):
    z1, z2, z3 = z
    rho = params['rho']
    sigma = params['sigma']
    beta = params['beta']

    dz1 = sigma * (z2 - z1)
    dz2 = z1 * (rho - z3) - z2
    dz3 = z1 * z2 - beta * z3

    return [dz1, dz2, dz3]

# some initial condition
znull = [1.0, 1.0, 1.0]
t_span = (0.0, 30.0)
t_eval = np.linspace(t_span[0], t_span[1], 20000)

sol = odeint(lorenz, znull, t_eval, args=(params,))

plt.plot(t_eval, sol)
    
```

```python
# plot in phase space
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:, 0], sol[:, 1], sol[:, 2])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Lorenz Attractor')
plt.show()
```
