# Notes on physics-informed EKF

## Damped oscillator 

So what you would usually do is you condition on your observation, this is what we already do in EKF-SINDy. However, we have *another* piece of information about the dynamics, we know that energy is also lost, and this loss depends on the the velocity and the dampening coefficients.

However, you are currently do the updates in a faulty manner. You first update based on the measurements, notice that the model is agnostic w.r.t. the energy loss, it just says "*ah, to assimilate this data I nudge my state and my parameters in this direction*". Why the parameters as well? They dictate the dynamics that have taken me from the previous state into my (slightly off) prediction.

However, I know that there's another link between the previous state and the current one, and that is that the energy is lost... 

The problem is that I already updated my state based only on matching the observation (rightly so). 

And know I ask: "ah, well, the energy difference should also be this, and your observation is know..."

But that's wrong, right? Because you should use the prediction given by your **current** estimate of the pseudo-observation. Would it not be better to use a **single** observation model?
